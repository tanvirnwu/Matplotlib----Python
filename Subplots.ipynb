# Subplots
In Python, a subplot refers to a way of organizing multiple plots in a single figure. It is a feature of the matplotlib library, which is a popular plotting library in Python.

A subplot is created by dividing the figure into a grid of rows and columns, and then assigning each plot to a specific cell in the grid. Each cell in the grid is called a subplot, and each subplot can contain its own plot.

To create subplots in matplotlib, you can use the subplot() function. This function takes three arguments: the number of rows in the grid, the number of columns in the grid, and the index of the subplot you want to create.

```python
import matplotlib.pyplot as plt

import pandas as pd
```

```python
# create a figure with two subplots arranged in a 1x2 grid
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

# plot some data on the first subplot
ax1.plot([1, 2, 3], [4, 5, 6])

# plot some data on the second subplot
ax2.plot([1, 2, 3], [6, 5, 4])

# plot some data on the third subplot
ax3.plot([6, 5, 4],[2,3,4])

# set the titles for each subplot
ax1.set_title('subplot 1')
ax2.set_title('subplot 2')
ax3.set_title('subplot 3')




# display the figure
plt.show()
```

```python
# create a figure with two subplots arranged in a 1x2 grid
fig, (ax1, ax2, ax3) = plt.subplots(3, 1)

# plot some data on the first subplot
ax1.plot([1, 2, 3], [4, 5, 6])

# plot some data on the second subplot
ax2.plot([1, 2, 3], [6, 5, 4])

# plot some data on the third subplot
ax3.plot([6, 5, 4],[2,3,4])

# set the titles for each subplot
ax1.set_title('subplot 1')
ax2.set_title('subplot 2')
ax3.set_title('subplot 3')


# display the figure
plt.show()
```

```python
plt.style.use('seaborn')

data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/10-Subplots/data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')
ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax.legend()

ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')
```
