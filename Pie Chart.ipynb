# Pie Chart

```python
from matplotlib import pyplot as plt
```

```python
slices= [60,40]
labels= ['Sixty','Fourty']
plt.tight_layout()
plt.title('My Pie Chart')

plt.pie(slices, labels=labels)
plt.legend()
```

```python
slices= [60,40, 20 ,30]
labels= ['Sixty','Fourty','Extra1','Extra2']
#colors=['blue','red','yellow','green']
colors=['#72CF81','#5D5E5E','#FF8C12','#28A9AD']

plt.tight_layout()
plt.title('My Pie Chart')


plt.pie(slices, labels=labels, colors=colors, wedgeprops={ 'edgecolor':'white'})
plt.legend()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
colors=['#72CF81','#5D5E5E','#FF8C12','#28A9AD','#6E9C2F']
explode = [0,0,0,0.1,0]


plt.tight_layout()
plt.title('My Pie Chart')

plt.pie(slices, labels=labels, colors=colors, explode= explode, shadow=True,
        startangle=90, autopct='%1.1f%%',
        wedgeprops={ 'edgecolor':'white'})

plt.legend()
```
