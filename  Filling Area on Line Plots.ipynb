# Filling Area on Line Plots

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv')
```

```python
data.head(2)
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='#1970E9', label='Python Devs')

plt.legend()
plt.title("Median Salary(USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Salary (USD)')
plt.tight_layout()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='#1970E9', label='Python Devs')

#overall_median = 57287
plt.fill_between(ages, py_salaries, alpha=0.30)

plt.legend()
plt.title("Median Salary(USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='#1970E9', label='Python Devs')

overall_median = 57287
plt.fill_between(ages, py_salaries,overall_median, alpha=0.30)

plt.legend()
plt.title("Median Salary(USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='#1970E9', label='Python Devs')

overall_median = 57287
plt.fill_between(ages, py_salaries,overall_median, where=(py_salaries>overall_median),
                 interpolate=True, alpha=0.30)

plt.legend()
plt.title("Median Salary(USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='#1970E9', label='Python Devs')

overall_median = 57287
plt.fill_between(ages, py_salaries,overall_median, where=(py_salaries > overall_median),
                 interpolate=True, alpha=0.30)

plt.fill_between(ages, py_salaries,overall_median, where=(py_salaries <= overall_median),
                 interpolate=True, color="#FF5033", alpha=0.30)

plt.legend()
plt.title("Median Salary(USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='#1970E9', label='Python Devs')

overall_median = 57287
plt.fill_between(ages, py_salaries,dev_salaries, 
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.30, label='Above Avg')

plt.fill_between(ages, py_salaries,dev_salaries, 
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color="#FF5033", alpha=0.30, label='Below Avg')

plt.legend()
plt.title("Median Salary(USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
```
